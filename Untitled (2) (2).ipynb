```python
import pandas as pd
#Importando o pyplot
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('./LAB02/breast-cancer.csv')
tumorsize = dataset.sort_values('tumor-size')
idade =  dataset.sort_values('idade')
degmaling = dataset.sort_values('deg-maling')
breast = dataset.sort_values('breast')
menopause = dataset.sort_values('menopause')
tumorsize.head()
```

```python
moda_class = dataset['class'].mode()
moda_idade = dataset['idade'].mode()
moda_menopause = dataset['menopause'].mode()
moda_tumor_size = dataset['tumor-size'].mode()
moda_invnodes = dataset['inv-nodes'].mode()
moda_nodes_caps = dataset['nodes-caps'].mode()
moda_degmaling = dataset['deg-maling'].mode()
moda_breast = dataset['breast'].mode()
moda_breastquad = dataset['breast-quad'].mode()
moda_irradiat = dataset['irradiat'].mode()

print('class', moda_class)
print('idade', moda_idade)
print('c', moda_menopause)
print('invnodes',moda_invnodes)
print('tumor_size',moda_tumor_size)
print('nodes_caps',moda_nodes_caps)
print('degmaling',moda_degmaling)
print('breast', moda_breast)
print('breast-quad', moda_breastquad)
print('irradiat', moda_irradiat)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(tumorsize['tumor-size'],idade['idade'])
plt.xlabel('Tamanho do tumor')
plt.ylabel('idade dos pacientes')
plt.title('Relação entre o tamanho de tumor com a idade do paciente')
plt.savefig('annual-real-gnp-us-1909-to-1970.png')

plt.show()
```
